# Visualización de rotaciones
Se muestran visualmente dos identidades de rotaciones en la esfera de Bloch

```python
from qiskit import *
import numpy as np
import matplotlib.pyplot as plt

from qiskit.visualization import *

plot_bloch_vector([0,0,1])
pi=np.pi
```

## $R_x(t)=HR_z(t)H$:

Hay dos qubits, a uno ($q_1$) se le aplica la rotación $R_x$ directamente, al otro ($q_0$) se le aplica la secuencia.

```python
qc=QuantumCircuit(2)
#Posicionamos en estado en una posición no trivial en la esfera
qc.rx(pi/4, [0,1])
#qc.ry(-pi/2,[0,1])

backend = Aer.get_backend('statevector_simulator')
out = execute(qc,backend).result().get_statevector()
plot_bloch_multivector(out)
```

```python
qc.h(0)
qc.rx(pi/2,1)
out = execute(qc,backend).result().get_statevector()
plot_bloch_multivector(out)
```

```python
qc.rz(pi/2,0)
out = execute(qc,backend).result().get_statevector()
plot_bloch_multivector(out)
```

```python
qc.h(0)
out = execute(qc,backend).result().get_statevector()
plot_bloch_multivector(out)
```


## Reset y hacemos $R_y(t)=R_z(\pi/2)R_x(t)R_z(-\pi/2)$

Nuevamente a q[0] se le aplica la secuencia, a q[1] se le aplica la rotación directamente.

```python
qc.reset([0,1])
qc.h([0,1])
qc.ry(-pi/6, [0,1])
out = execute(qc,backend).result().get_statevector()
plot_bloch_multivector(out)
```

```python
qc.rz(-pi/2,0)
qc.ry(pi/3,1)
out = execute(qc,backend).result().get_statevector()
plot_bloch_multivector(out)
```

```python
qc.rx(pi/3,0)
out = execute(qc,backend).result().get_statevector()
plot_bloch_multivector(out)
```

```python
qc.rz(pi/2,0)
out = execute(qc,backend).result().get_statevector()
plot_bloch_multivector(out)
```

En ambos casos, ambos qubits terminan en el mismo estado (visualmente), mostrando las identidades en rotaciones.
